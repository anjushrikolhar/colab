```python
import pandas as pd
iris=pd.read_csv('/content/Iris.csv')
iris.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(iris['PetalLengthCm'], iris['PetalWidthCm'])
plt.title('scatter plot')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(iris['PetalLengthCm'], bins=5, color='green', edgecolor='black')
plt.title("Distribution of Petal Length")
plt.xlabel("PetalLengthCm")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('/content/Iris.csv')
Species_counts = iris['Species'].value_counts()
plt.bar(Species_counts.index, Species_counts.values)
plt.title('Number of Flowers by Species')
plt.xlabel('Species')
plt.ylabel('frequency')
plt.show()
print(Species_counts)
```

```python
import matplotlib.pyplot as plt
Species = iris['Species'].unique()
data = [iris[iris['Species'] == s]['PetalLengthCm'] for s in Species]
plt.boxplot(data, labels=Species)
plt.title("Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.show()
```
